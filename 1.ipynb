```python
from groq import Groq
import os
YOUR_GROQ_API_KEY=os.environ['GROQ_API_KEY']
client = Groq(api_key=YOUR_GROQ_API_KEY)
```

```python
request='you are an strategic trader with 20 years of experience in crypto trading  understanding every patterns and atrategiescome up with a most accurate trading strategy for futures and option in crypto with 100 percent accuracy'
request+='answer the question'
message=[{'role':'user','content':request}]
```

```python
response=client.chat.completions.create(model='llama-3.1-8b-instant',messages=message)

question=response.choices[0].message.content
question
```
